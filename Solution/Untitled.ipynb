```python
%matplotlib inline  
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
easy_value= pd.read_csv('./Easy Value.csv')
easy_policy= pd.read_csv('./Easy Policy.csv')


ax=easy_value.iloc[:, [0,2]].plot(x=0, title="Rewards",figsize=(8,5))
easy_policy.iloc[:, [0,2]].plot(ax=ax,x=0, title="Rewards",figsize=(8,5))

ax.set_xlabel("# of iterations",fontsize=12)
ax.set_ylabel("Rewards",fontsize=12)
ax.set_title("Rewards of VI & PI",fontsize=12)
ax.legend(["Value Iteration", "Policy Iteration"]);



ax=easy_value.iloc[:, [0,3]].plot(x=0, title="Steps",figsize=(8,5))
easy_policy.iloc[:, [0,3]].plot(ax=ax,x=0, title="Steps",figsize=(8,5))

ax.set_xlabel("# of iterations",fontsize=12)
ax.set_ylabel("Steps to Goal",fontsize=12)
ax.set_title("Steps to Goal of VI & PI",fontsize=12)
ax.legend(["Value Iteration", "Policy Iteration"]);




ax=easy_value.iloc[:, [0,4]].plot(x=0, title="Convergence",figsize=(8,5))
easy_policy.iloc[:, [0,4]].plot(ax=ax,x=0, title="Convergence",figsize=(8,5))

ax.set_xlabel("# of iterations",fontsize=12)
ax.set_ylabel("Convergence",fontsize=12)
ax.set_title("Convergence of VI & PI",fontsize=12)
ax.legend(["Value Iteration", "Policy Iteration"]);




ax=easy_value.iloc[:, [0,1]].plot(x=0, title="Time",figsize=(8,5))
easy_policy.iloc[:, [0,1]].plot(ax=ax,x=0, title="Time",figsize=(8,5))

ax.set_xlabel("# of iterations",fontsize=12)
ax.set_ylabel("Time",fontsize=12)
ax.set_title("Accumulated Time of VI & PI",fontsize=12)
ax.legend(["Value Iteration", "Policy Iteration"]);



print easy_value
# easy_value.iloc[:, 0:2].plot(ax=ax,x=0, title="SSE",figsize=(8,5))
```
